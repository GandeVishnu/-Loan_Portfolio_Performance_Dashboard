```python
# Function to clean emp_length column
def clean_emp_length(x):
    if pd.isna(x):
        return None
    x = str(x).strip()
    if x == "< 1 year":
        return 0.5   
    elif x == "10+ years":
        return 10
    elif "year" in x:
        return int(x.split()[0])
    return None

# Apply function to create new numeric column
dt["emp_length_num"] = dt["emp_length"].apply(clean_emp_length)



print("âœ… Cleaned data saved to 'cleaned_emp_length.xlsx'")
print(dt.head())
```
